# Предобработка

Посмотрим на уникальные значения в столбцах `has_Green_id`, `has_Black_id`, `has_Blue_id`.

```python
print(pd.unique(data.has_Green_id))
print(pd.unique(data.has_Black_id))
print(pd.unique(data.has_Blue_id))
```

Напишем функцию которая заменяет все значения большие 1 на единицу, а 0 оставляет нулём.

```python
def replacer(n: int) -> int:
    if n > 1:
        return 1
    
    return 0
```

Исправим значения признаков `has_Green_id`, `has_Black_id`, `has_Blue_id`, применив вышенаписанную функцию к этим столбцам.

```python
data.has_Green_id = data.has_Green_id.apply(replacer)
data.has_Black_id = data.has_Black_id.apply(replacer)
data.has_Blue_id = data.has_Blue_id.apply(replacer)
```

Переименуем столбец `active_Red`  в `has_Red_id`, так как по принимаемым значениям и названию можно думать что это он и есть. Также исправим опечатки в `Green_gvm_total` и `Black_gvm_total`.

```python
data.rename(columns={'active_Red': 'has_Red_id', 
                     'Green_gvm_total': 'Green_gmv_total', 
                     'Black_gvm_total': 'Black_gmv_total'}, 
            inplace=True)

data.tail(5) # смотрим на последние 5 наблюдений
```

Посмотрим на уникальные значения признака `Blue_activity_count`.

```python
set(data.Blue_activity_count)
```

Видно что там есть как строки, так и даты. Напишем функцию переводящую строки в целые числа, а дату заменяет на 0, потому что если даты переводить в числа, то получатся либо дробные числа (чего быть не может по сути признака), а если переводить в Unix формат (количество секунд от первого января 1970 года) то получится черезчур большое и нереалистичное число.

```python
def converter(el) -> int:
    if isinstance(el, str):
        return int(el.split('.')[0])

    return 0
```

```python
data.Blue_activity_count = data.Blue_activity_count.apply(converter)
```

# Предсказание признаков

Так как по некоторым компаниям отсутствует часть их призаков, попробуем спрогнозировать эти значения с помощью машинного обучения.

Значения признака `comp_activity_count` есть только у компании `Blue`, поэтому предсказывать этот признак будем по данным о компании `Blue`, на основании признаков которые есть у всех компаний.

```python
X_Red = data[['has_Red_id', 'Red_android', 'Red_apple', 'Red_web', 'is_msk']]
X_Green = data[['has_Green_id', 'Green_android', 'Green_apple', 'Green_web', 'is_msk']]
X_Black = data[['has_Black_id', 'Black_android', 'Black_apple', 'Black_web', 'is_msk']]

X_Blue = data[['has_Blue_id', 'Blue_android', 'Blue_apple', 'Blue_web', 'is_msk']]
y_Blue = data.Blue_activity_count
```

Разделим данные на тестовую и тренировочную выборку.

```python
from sklearn.model_selection import train_test_split

X_Blue_train, X_Blue_test, y_Blue_train, y_Blue_test = train_test_split(X_Blue, y_Blue, test_size=0.25, random_state=42)
```

По своей сути это признак принимающий натуральные значения, поэтому будем строить регрессию с L1 регуляризацией (коеф. регуляризации возьмём 0.0001), в качестве функционала качества возьмём MSE, а затем просто округлим предсказанные значения если они окажутся не целочисленными.

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error

ridge_model1 = Lasso(0.0001)

ridge_model1.fit(X_Blue_train, y_Blue_train)
y_Blue_pred = ridge_model1.predict(X_Blue_test).astype(int)

print(mean_squared_error(y_Blue_pred, y_Blue_test))
```

Получили довольно хорошоее значение MSE, поэтому будем применять эту модель для предсказания этого призанка у других компаний.

```python
data['Red_activity_count'] = ridge_model1.predict(X_Red).astype(int)
data['Green_activity_count'] = ridge_model1.predict(X_Green).astype(int)
data['Black_activity_count'] = ridge_model1.predict(X_Black).astype(int)
```

Аналогичным образом, но уже на основании данных о компаниях `Green` и `Black` предскажем признак `comp_gvm_order_count` для компаний `Red` и `Blue`.
